## 전처리 및 분석

### 운송금액 전처리 > 숫자변환 > 히스토그램

```python
# 전체 row 출력
#pd.set_option('display.max_rows', None)
#pd.options.display.max_rows = 60
```

```python
# 운송금액
df.ACT_DTL13.value_counts().sort_values(ascending=False)
```

```python
# 문자 대체 ","->""
# df.loc[df.ACT_DTL13 =="56,000", 'ACT_DTL13']= 56000
df.ACT_DTL13=df.ACT_DTL13.str.replace(",","")
```

```python
# 이상값처리 : ';' 기준 분리하고, 정수로 만들어 첫번째 값을 'ACT_DTL16' 에 저장
df['ACT_DTL16']=df.loc[df.ACT_DTL13.notnull(), 'ACT_DTL13'].str.split(';').str[0].astype(int)
```

```python
df.ACT_DTL16.value_counts().sort_index(ascending=False)
```

### EDA - 차종, 무게

```python
# 차종, 무게의 값을 보기
df.ACT_DTL7.value_counts().sort_values(ascending=False)
```

```python
# 이상값처리 : ';' 기준 분리하여 첫번째 값을 신규필드('car') 에 저장
df['car']=df.loc[df.ACT_DTL7.notnull(), 'ACT_DTL7'].str.split(';').str[0]
```

```python
# 이상값처리 : ';' 기준 분리하여 두번째 값을 신규필드('ton') 에 저장
df['ton']=df.loc[df.ACT_DTL7.notnull(), 'ACT_DTL7'].str.split(';').str[1]
```

```python
df.car.value_counts().sort_values(ascending=False)
```

```python
# groupby count 를 하여 데이터 형태를 보기
df0=df.groupby(['car','ton'],as_index=False)['ACT_DTL7'].count()
df0.info()
df0
```

```python
# 시각화를 위한 히트맵으로 만들기 위해 피벗테이블로 만들기
dfp=df0.pivot(index='car',columns='ton',values='ACT_DTL7')
dfp.drop(columns=['','0.0'],inplace=True)  # 첫,두번째 컬럼 삭제
dfp.drop(dfp.index[[0]],inplace=True)      # 첫번째 행 삭제
dfp=dfp.fillna(0).astype(int)              # 빈값을 '0'으로 채우고 값을 '정수'로 변경
dfp
```
